### Q1. What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a Python data visualization library used for creating static, animated, and interactive visualizations in Python. It is one of the most popular data visualization libraries available in Python and is widely used in the scientific and engineering communities for creating publication-quality plots and charts.

Matplotlib is used for a variety of purposes, such as:

Exploring data and visualizing relationships and patterns within data.
Communicating insights and findings to others through data visualizations.
Building custom visualizations to support scientific research, engineering analysis, or business reporting.
Some of the common plots that can be created using the Pyplot module of Matplotlib are:

Line plot: A line plot is a basic plot that is used to display data as a series of points connected by straight lines.

Scatter plot: A scatter plot is used to visualize the relationship between two continuous variables by plotting individual data points on a two-dimensional plane.

Bar plot: A bar plot is used to compare the values of different categories using rectangular bars of lengths proportional to the values being compared.

Histogram: A histogram is used to visualize the distribution of a single continuous variable by dividing the data into a set of bins and plotting the frequency of observations in each bin.

Pie chart: A pie chart is used to show the proportion of each category in a dataset, where each category is represented as a slice of a circular pie.

The library is highly customizable, and its extensive documentation provides users with the flexibility to create any type of visualization they need.

### Q2. What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

#### import numpy as np
#### np.random.seed(3)
#### x = 3 + np.random.normal(0, 2, 50)
#### y = 3 + np.random.normal(0, 2, len(x))

A scatter plot is a type of plot used to visualize the relationship between two continuous variables. It is commonly used to explore correlations and patterns in data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y)

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter plot of X vs Y')

# Display the plot
plt.show()
```

We first generate the data for x and y using NumPy's random.normal() method with mean 3 and standard deviation 2. We then create a scatter plot using plt.scatter() method of Matplotlib's Pyplot module.

Finally, we add labels to the x and y axes using plt.xlabel() and plt.ylabel() respectively, and add a title to the plot using plt.title(). We display the plot using plt.show() method. The resulting plot will show the relationship between the x and y variables as a scatter plot with individual data points plotted as circles.

### Q3. Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

#### import numpy as np
#### For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
#### For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
#### For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
#### For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in Matplotlib is used to create multiple plots in the same figure. It allows us to create a grid of subplots, where each subplot can display a different plot or visualization.

The subplot() function takes three arguments: nrows, ncols, and index, which specify the number of rows and columns in the grid, and the index of the subplot to create. The index starts from 1 and increases row-wise, left to right.

Example

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(nrows=2, ncols=2)

# Plot the four lines
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# Add titles to each subplot
axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

# Display the plot
plt.show()
```

We first generate the data for the four lines using NumPy's array() method. We then create a 2x2 grid of subplots using the subplot() function and store the subplots in the axs variable.

Next, we plot the four lines using the plot() method of each subplot, and add titles to each subplot using the set_title() method.

Finally, we display the plot using the show() method. The resulting plot will show four line plots, each in its own subplot within the 2x2 grid.

### Q4. What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

#### import numpy as np
#### company = np.array(["Apple", "Microsoft", "Google", "AMD"])
#### profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a visualization tool that displays categorical data with rectangular bars. The height or length of each bar represents the value or frequency of the corresponding category. Bar plots are commonly used to compare different categories or to show changes in a category over time.

A horizontal bar plot is a type of bar plot where the bars are drawn horizontally instead of vertically. This type of plot is useful when the category names are long or when there are many categories to display.

Example:

```python
import numpy as np
import matplotlib.pyplot as plt

# Create data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.bar(company, profit)

# Add labels and title
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Annual profit of tech companies')

# Show the plot
plt.show()
```

This code creates a vertical bar plot using the bar() function of Matplotlib. The bar() function takes two arguments: the categories to be plotted (company), and the values for each category (profit).

We then add labels and a title to the plot using the xlabel(), ylabel(), and title() functions. Finally, we display the plot using the show() function.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a horizontal bar plot
plt.barh(company, profit)

# Add labels and title
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.title('Annual profit of tech companies')

# Show the plot
plt.show()
```

This code is similar to the previous example, except we use the barh() function instead of bar(). We also switch the order of the labels on the x and y axes, and adjust the labels and title accordingly.

### Q5. What is a box plot? Why is it used? Using the following data plot a box plot.

#### box1 = np.random.normal(100, 10, 200)
#### box2 = np.random.normal(90, 20, 200)

A box plot is a graphical representation of a data set using five summary statistics: minimum, first quartile, median, third quartile, and maximum. It is also called a box-and-whisker plot because it consists of a box (representing the second and third quartiles), whiskers (representing the minimum and maximum values), and sometimes outliers (points outside the whiskers).

Box plots are commonly used to visualize the distribution of a data set, including its central tendency, spread, and skewness. They are especially useful when comparing multiple data sets or when dealing with large data sets with many outliers.

Example:

```python
import numpy as np
import matplotlib.pyplot as plt

# Create data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]

# Create a box plot
plt.boxplot(data)

# Add labels and title
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.ylabel('Value')
plt.title('Box plot example')

# Show the plot
plt.show()
```

This code creates a box plot using the boxplot() function of Matplotlib. The boxplot() function takes a list of data sets to be plotted (data in this case).

We then add labels and a title to the plot using the xticks(), ylabel(), and title() functions. The xticks() function is used to set the labels for the x-axis ticks. Finally, we display the plot using the show() function.

This code generates two random data sets box1 and box2. We then put them in a list data and pass it to the boxplot() function. The resulting box plot compares the distribution of box1 and box2. The boxes represent the interquartile range (IQR), which contains 50% of the data. The whiskers extend to the minimum and maximum values, excluding any outliers. If there are outliers, they are shown as individual points. In this example, the two boxes have different medians and spreads, and box 2 has more outliers than box 1.
