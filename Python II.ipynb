### Q1. Write a program to accept percentage from the user and display the grade according to the following criteria: 



```python
percentage = float(input("Enter your percentage: "))

if percentage > 90:
    grade = 'A'
elif percentage >80 and percentage <=90:
    grade = 'B'
elif percentage >= 60 and percentage <=80:
    grade = 'C'
else:
    grade = 'D'

print("Your grade is:", grade)
```

### Q2. Write a program to accept the cost price of a bike and display the road tax to be paid according to the following criteria:



```python
costprice = int(input("Enter the cost price of a bike: "))

if costprice > 100000:
    tax = 0.15*costprice
elif costprice > 50000 and costprice <= 100000:
    tax = 0.1*costprice
else:
    tax = 0.05*costprice
    
print("The road tax to be paid is:", tax)
```

### Q4. Check how many times a given number can be divided by 3 before it is less than or equal to 10

```python
num = int(input("Enter a number: "))
count = 0

while num > 10:
    num /= 3
    count += 1

print("The given number can be divided by 3", count, "times before it becomes less than or equal to 10.")
```

### Q5. Why and when to use while loop in Python. Give a detailed description with example

The while loop is a fundamental control structure in Python that allows us to execute a block of code repeatedly as long as a specified condition is true. The while loop is useful when we want to repeat a block of code an unknown number of times, or when we want to repeat a block of code until a certain condition is met.

Example

```python
# Initialize the counter variable
count = 0

# Start a while loop that executes as long as count < 10
while count < 10:
    # Check if the current number is even
    if count % 2 == 0:
        # Print the current number
        print(count)
    # Increment the counter variable
    count += 1
```

In general, we use a while loop in Python when we need to repeat a block of code until a certain condition is met, and we don't know ahead of time how many times we need to repeat the code. Some common use cases for while loops include:

Reading input from a user until they enter a valid value

Repeating a simulation until a certain outcome is achieved

Processing data until a certain condition is met (e.g. processing a file until the end of the file is reached)






### Q6. Use nested while loop to print 3 different patterns

```python
Pattern:
    
*
**
***
****
*****
```

```python
i = 1

while i <= 5:
    j = 1
    while j <= i:
        print("*", end="")
        j += 1
    print()
    i += 1
```

In this example, we use two nested while loops to print a pattern of asterisks. The outer loop iterates over the rows of the pattern (i.e. from 1 to 5), while the inner loop iterates over the columns of the pattern (i.e. from 1 to the current row number).

```python
Pattern:

    *
   **
  ***
 ****
*****
```

```python
i = 1

while i <= 5:
    j = 1
    while j <= (5 - i):
        print(" ", end="")
        j += 1
    k = 1
    while k <= i:
        print("*", end="")
        k += 1
    print()
    i += 1
```

In this example, we use three nested while loops to print a pattern of asterisks and spaces. The outer loop iterates over the rows of the pattern (i.e. from 1 to 5), while the inner loop iterates over the columns of the left-hand side of the pattern (i.e. from 1 to the number of spaces needed to align the asterisks on the right-hand side of the pattern). The third loop iterates over the columns of the right-hand side of the pattern (i.e. from 1 to the current row number).

```python
Pattern:

    *
   ***
  *****
 *******
*********
 *******
  *****
   ***
    *
```

### Q7. Reverse a while loop to display numbers from 10 to 1
